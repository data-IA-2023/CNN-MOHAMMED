```python
import os
from collections import defaultdict
import matplotlib.pyplot as plt

# Chemin du dossier contenant les sous-dossiers de photos
dossier_parent = "PetImages"

# Dictionnaire pour stocker le nombre de photos par type d'animal
nombre_photos_par_type = defaultdict(int)

# Parcourir tous les sous-dossiers
for dossier_animal in os.listdir(dossier_parent):
    chemin_dossier_animal = os.path.join(dossier_parent, dossier_animal)
    # Vérifier si le chemin est un dossier
    if os.path.isdir(chemin_dossier_animal):
        # Compter le nombre de fichiers dans le dossier (supposant que ce sont des photos)
        nombre_photos = len(os.listdir(chemin_dossier_animal))
        # Récupérer le type d'animal à partir du nom du dossier
        type_animal = dossier_animal.split("_")[-1]  # Supposant que le nom du dossier est dans le format "animal_type"
        # Ajouter le nombre de photos au type d'animal correspondant
        nombre_photos_par_type[type_animal] += nombre_photos

# Créer le diagramme à barres
types_animaux = list(nombre_photos_par_type.keys())
nombre_photos = list(nombre_photos_par_type.values())

plt.bar(types_animaux, nombre_photos, color='skyblue')
plt.xlabel('Type d\'animal')
plt.ylabel('Nombre de photos')
plt.title('Nombre de photos par type d\'animal')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
